```python
import osmnx as ox
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=[10,8])
stras = ox.geocode_to_gdf("Strasbourg, France")
ax = ox.project_gdf(stras).plot()
_ = ax.axis("off")
plt.title("Strasbourg")
```

```python
plt.figure(figsize = [10,5])
hdb = ox.geocode_to_gdf("Heidelberg, Germany")
ax = ox.project_gdf(hdb).plot()
_ = ax.axis("off")
plt.title("Heidelberg")
```

```python
ox.plot_graph(ox.graph_from_place("Strasbourg, France"))
```
