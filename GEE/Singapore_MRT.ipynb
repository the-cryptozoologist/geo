```python
import folium
```

```python
dotmap = folium.Map(location=[1.3521, 103.8198], zoom_start=10)
dotmap
```

```python
for lat, lon, boro, nbh in zip(df["Latitude"], df["Longitude"], df["STN_NAME"], df["COLOR"]):
  label = f"{nbh}, {boro}"
  label = folium.Popup(label, parse_html=True)
  folium.CircleMarker(
      [lat,lon],
      radius=5,
      popup=label,
      color="blue",
      fill=True,
      fill_color="pink",
      fill_opacity=0.7,
      parse_html=False).add_to(dotmap)
dotmap
```

```python
sg_lat=1.3521
sg_lon=103.8198

DFW_Map = folium.Map(
    location=[sg_lat, sg_lon],
    zoom_start=10)

lats=list(df.Latitude)
lons=list(df.Longitude)
labels=list(df.COLOR)

for lat, lon, label in zip(lats, lons, labels):
  if label == "RED":
    folium.Marker(
        location = [lat,lon],
        popup = label,
        icon = folium.Icon(color="red", icon="circle")
    ).add_to(DFW_Map)
  elif label == "GREEN":
    folium.Marker(
        location = [lat,lon],
        popup = label,
        icon = folium.Icon(color="green", icon ="circle")
    ).add_to(DFW_Map)
  elif label == "YELLOW":
    folium.Marker(
        location = [lat,lon],
        popup = label,
        icon = folium.Icon(color="yellow", icon ="circle")
    ).add_to(DFW_Map)
  elif label == "PURPLE":
    folium.Marker(
        location = [lat,lon],
        popup = label,
        icon = folium.Icon(color="purple", icon ="circle")
    ).add_to(DFW_Map)
  elif label == "BLUE":
    folium.Marker(
        location = [lat,lon],
        popup = label,
        icon = folium.Icon(color="blue", icon ="circle")
    ).add_to(DFW_Map)
  else:
    folium.Marker(
        location = [lat,lon],
        popup = label,
        icon = folium.Icon(color="grey", icon="circle")
    ).add_to(DFW_Map)

DFW_Map
```
